```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
def sumPowell(x):
  sumY = 0
  for index, xValue in enumerate(x, 1):
    #print(index, xValue)
    absX = np.absolute(xValue) 
    powerX = np.power(absX, index + 1)
    sumY += powerX

  #print(sumY)   
  return sumY
```

```python
N = 10000
D = 2
D_lower = np.ones((1,D)) * - 1
D_upper = np.ones((1,D)) * 1
```

```python
X = pd.DataFrame(np.random.uniform(low=D_lower, high=D_upper, size=(N,D)), columns=['x1','x2'])
```

```python
Y = np.empty((0,1))
```

```python
for n in range(N):  
  #print(X.iloc[n]) 
  y = sumPowell(X.iloc[n])
  #break  
  Y = np.vstack([Y,y])
```

```python
Y
```

```python
Y = pd.DataFrame(Y,columns=['f'])
```

```python
fig = go.Figure(layout=dict(scene=dict(xaxis_title='x1', yaxis_title='x2', zaxis_title='f')))

fig.add_scatter3d(x=X.x1, y=X.x2, z=Y.f, mode='markers', marker=dict(color=-Y.f))

fig.show()
```
